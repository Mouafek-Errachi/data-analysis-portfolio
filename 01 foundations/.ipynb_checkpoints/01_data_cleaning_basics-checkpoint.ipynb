# Data Cleaning Basics
## Objective: Clean and prepare the tips dataset for analysis.
### Preparing Dataset
#### 1-Importing data

```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)
```

#### 2-Viewing Data

```python
df.info()
```

#### 3-Checking Missing Values

```python
df.isnull().sum()
```

#### 4-Renaming Data in snake_case

```python
df.columns=(
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ","_")
)
```

#### 5-Adding "tip_percent" column

```python
df["tip_percent"] = df["tip"] / df["total_bill"] * 100
```

### Analysing "tip_percent" Dataset
#### 1- Grouping by Day

```python
df.groupby("day")["tip_percent"].mean()
```

#### 2- Viewing "tip_percent" Histogram

```python
df["tip_percent"].hist()
```
