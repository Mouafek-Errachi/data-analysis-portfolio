# Time Series Analysis Basics
## Objective: Understand trends and patterns in sequential performance data.


```python
import pandas as pd
import matplotlib.pyplot as plt

laps = list(range(1, 21))
lap_time = [
    90.2, 90.1, 90.0, 90.1, 90.2,
    90.4, 90.6, 90.8, 91.0, 91.2,
    91.4, 91.6, 91.8, 92.0, 92.2,
    92.4, 92.6, 92.8, 93.0, 93.2
]

df = pd.DataFrame({
    "lap": laps,
    "lap_time": lap_time
})
plt.plot(df["lap"], df["lap_time"])
plt.xlabel("Lap")
plt.ylabel("Lap Time (s)")
plt.title("Lap Time Evolution Over a Stint")
plt.show()
```

- This time series shows the evolution of lap time over a stint. 
- Lap times decrease slightly during the first few laps, likely due to the car reaching optimal operating conditions. 
- After lap 3, lap times increase approximately linearly, indicating a loss of performance over the stint. 
- This trend is consistent with tire degradation accumulating over time.
